```python
forecast = nixtla_client.forecast(
    df_filtered,
    time_col='Date',        
    target_col='GDP_M',               
    freq='Q',               
    h=12,             
)
```

```python
std_dev_gdp = df_filtered['GDP_M'].std()
moe = std_dev_gdp
```

```python
forecast['lower'] = forecast['TimeGPT'] - (1.645 * moe)  
forecast['upper'] = forecast['TimeGPT'] + (1.645 * moe) 
```

```python
forecast_df = pd.DataFrame(forecast)
```

```python
last = df_filtered['Date'].iloc[-1]
```

```python
connector = df_filtered.iloc[-1] 
forecast_df = pd.concat([connector, forecast_df], ignore_index=True)
```

```python
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'])
forecast_df['Date'] = pd.to_datetime(forecast_df['Date'])
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))
plt.plot(df_filtered['Date'], df_filtered['GDP_M'], label='Actual GDP', color='blue')
plt.plot(forecast_df['Date'], forecast_df['TimeGPT'], label='Forecasted GDP', color='orange')
plt.fill_between(forecast_df['Date'], 
                 forecast_df['lower'],  
                 forecast_df['upper'],  
                 color='pink', alpha=0.5, label='Confidence Interval (90%)')
plt.title('GDP Data + Forecast')
plt.xlabel('Date')
plt.ylabel('GDP($dollars)')
plt.legend(loc='upper left')
plt.show()
```

```python
forecast_df.drop(columns=[0], inplace=True)
forecast_df = forecast_df.dropna(subset=['Date', 'TimeGPT'])
forecast_df
```

```python
df_train = df_filtered[(df_filtered['Date'] >= '1999-03-31') & (df_filtered['Date'] <= '2019-12-31')]
df_test = df_filtered[(df_filtered['Date'] >= '2020-01-01') & (df_filtered['Date'] <= '2024-12-31')]
```

```python
forecast2 = nixtla_client.forecast(
    df_train,
    time_col='Date',
    target_col='GDP_M',
    freq='Q',
    h=20  
)
```

```python
forecasted_values = forecast2[['Date', 'TimeGPT']]
```

```python
df_comparison = df_test.merge(forecasted_values, on='Date', suffixes=('_actual', '_forecast'))
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(df_comparison['GDP_M'], df_comparison['TimeGPT'])
print(f"Mean Absolute Error(MAE) for 2020-2024: {mae}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_comparison['Date'], df_comparison['GDP_M'], label='Actual GDP', color='blue')
plt.plot(df_comparison['Date'], df_comparison['TimeGPT'], label='Forecasted GDP', color='orange')

plt.title('Actual vs Forecasted GDP')
plt.xlabel('Date')
plt.ylabel('GDP($)')
plt.legend()
plt.grid()
plt.show()
```

```python
df_comparison['residuals'] = df_comparison['GDP_M'] - df_comparison['TimeGPT']

plt.figure(figsize=(10, 6))
plt.plot(df_comparison['Date'], df_comparison['residuals'], marker='o', linestyle='-', color='red')
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Residuals (Actual vs Forecast)')
plt.xlabel('Date')
plt.ylabel('Residual (GDP Actual - Forecast)')
plt.show()
```
